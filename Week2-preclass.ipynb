```python
# Read in data
URL = 'https://raw.githubusercontent.com/allisonhorst/palmerpenguins/main/inst/extdata/penguins.csv'
penguins = pd.read_csv(URL)

penguins.head()
```

```python
#1. 

penguins[['bill_length_mm','bill_depth_mm']].plot(kind='box',
                                title='Penguin flipper lengths',
                                ylabel='Length (mm)',
                                grid=True,
                                color = 'goldenrod')
```

```python
# 2. Create a simple histogram of the flipper length of female gentoo penguins.
female_gentoo = penguins[(penguins['sex'] == 'female') & (penguins['species'] == 'Gentoo')]

female_gentoo['flipper_length_mm'].plot(kind = 'hist',
                                       title = 'flipper length in female gentoo penguins',
                                       xlabel = 'flipper length (mm)',
                                       color = 'goldenrod')
```

```python
penguins.loc[(penguins['sex'] == 'female') & (penguins['species'] == 'Gentoo'), 'flipper_length_mm'].plot(kind = 'hist',
                                       title = 'flipper length in female gentoo penguins',
                                       xlabel = 'flipper length (mm)',
                                       color = 'goldenrod')
```
